```python
import csv
import sys


Train_Data = []
Title_Dict = {}
with open ("train.tsv",'r')as infile:
    title = infile.readline().replace('\n','').split('\t')
    for i in range(len(title)):
        Title_Dict[title[i]] = i
    print(Title_Dict)
    Datas = infile.readlines()
    for the_line in Datas:
        data_line = the_line.replace('\n','').split('\t')
        Train_Data.append(data_line)
    infile.close()    
```

```python
class Node :
    def __init__(self,node_name):
        self.name = node_name
        self.parent = None
        self.value = 1
        self.child = []
        
    def printTree(self,level=0):
        print('\t'*level + str(self.name) +" : "+str(self.value) )
        for child_node in self.child:
            if child_node != None:
                child_node.printTree(level+1)
```

```python
def searchCategories(Root, categories):
    Node_now = Root
    new_cate = []
    for i in range(3):
        if_find = 0
        for index in range(len(Node_now.child)):
            if categories[i] == Node_now.child[index].name:
                new_cate.append(index)
                Node_now = Node_now.child[index]
                if_find = 1
                break
        if if_find == 0:
            new_Node = Node(categories[i])
            Node_now.child.append(new_Node)
            new_cate.append(len(Node_now.child))
            Node_now = new_node
    return new_cate
```
